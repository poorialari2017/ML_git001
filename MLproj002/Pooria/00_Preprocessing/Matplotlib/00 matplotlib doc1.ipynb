```python
import matplotlib.pyplot as plt
import numpy as np
```

## first test

```python
# دقت کنید مقدار x  با y باید یکسان باشه

x = [1,2,3,4,5,6]
y = [2,6,3,10,50,12]
# ساخت نمودار اولیه
# اضافه کردن رنگ
plt.plot(x, y, color='red')

## اضافه کردن لیبل و تیتر
plt.xlabel('x', color='blue')
plt.ylabel('y', color='green')
plt.title('Plot title')
plt.show()
```

```python
## دایره کردن  نمودار ها

plt.plot(x,y,'o',color='yellow')
plt.xlabel('x', color='blue')
plt.ylabel('y', color='green')
```

```python
##  ستاره ای کردنشون
plt.plot(x,y,'*',color='yellow')
plt.xlabel('x', color='blue')
plt.ylabel('y', color='green')
```

```python
## s ستاره ای کردن یا دایره ای کردن و ووصل کردن

plt.plot(x,y,marker = 'o',color='yellow')
plt.title('Plot title')
plt.show()
```

### k نمودار میله ای بارچارت


```python
x = ['blue', 'green', 'red', 'yellow', 'purple']
y = [2,3,19,53,12]
plt.bar(x,y, color='red')
plt.xlabel('x', color='blue')
plt.ylabel('y', color='green')
plt.title('Plot title')
plt.xlabel('color')
plt.ylabel('number')
plt.show()
```

### . k نمودار دایره ای

```python
### autopct='%1.1f%% برای اینه که دقت درصد رو مشخص کنه


plt.pie(y,labels= x, autopct='%1.1f%%')

plt.show()
```

```python
### autopct='%1.1f%% برای اینه که دقت درصد رو مشخص کنه


plt.pie(y,labels= x, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'yellow', 'purple'])

plt.show()
```

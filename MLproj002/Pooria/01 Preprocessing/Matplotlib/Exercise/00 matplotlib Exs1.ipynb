```python
import numpy as np
import matplotlib.pyplot as plt
X = np.arange(0,11)
Y = 2*X + 1
plt.grid(True)
plt.title("Matplotlib")
plt.xlabel("X")
plt.ylabel("Y")
plt.plot(X, Y, label='y = 2x + 1', marker='o')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(0,361)
y = np.sin(X)
plt.plot(X, y)
plt.title('sin range')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_deg = np.arange(0,361)

x = np.deg2rad(x_deg)
y = np.sin(x)
plt.plot(x, y)
plt.title('sin range')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#x = np.arange(-5,5)
x = np.linspace(-5,5, 201)
y1 = x
y2 = x**2
plt.plot(x, y1,marker='o', label='y1',linewidth=2)
plt.plot(x, y2, linestyle='--', label='y2',linewidth=2)
plt.title('monhani')
plt.grid(True)
plt.xlim(-5, 5)
plt.ylim(-1, 26)
plt.legend()
plt.show()
```

```
frequency1 = 0
frequency2 = 0
frequency3 = 0
frequency4 = 0
frequency5 = 0
frequency6 = 0

for i in range(6_000_000):
    face = random.randint(1,6)
    if face == 1:
        frequency1 += 1
    elif face == 2:
        frequency2 += 1
    elif face == 3:
        frequency3 += 1
    elif face == 4:
        frequency4 += 1
    elif face == 5:
        frequency5 += 1
    else:
        frequency6 += 1


print(frequency1+frequency2+frequency3+frequency4+frequency5+frequency6)
```

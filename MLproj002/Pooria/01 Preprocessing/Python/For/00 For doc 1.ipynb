```python
grades = [16,14,18,6,15,20,17,17,5]
sum = 0
count = 0
for grade in grades:
    sum += grade
    count += 1
    print(grade, count, sep="-")

average = sum/count
print(f'Sums is : {sum}')
print(f'Counts is : {count}')
print(f'Average is : {average:0.2f}')

del sum, average, grades,count, grade
gc.collect()
```
